```python
import pandas as pd
import numpy as np

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTMs
from tensorflow.keras.layers import Dense
```

```python
message=pd.read_csv("https://raw.githubusercontent.com/SaurabhSRP/04-NLP-based-Projects/main/SMS%20Spam%20classification%20using%20LSTM/SMSSpamCollection",sep='\t',names=["lable","message"])
message.head()
```

```python
message.shape
```

```python
map={'ham':0,'spam':1}
message['lable']=message['lable'].map(map)
```

#**Feature Engineering**
* Stop words
* lemmatization
* Word Embedding


```python
import nltk
import re
from nltk.corpus import stopwords

nltk.download('stopwords')

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0, len(message)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', message['message'][i])
    review = review.lower()
    review = review.split()
    
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus[0]
```

```python
voc_size=5000
onehot_repr=[one_hot(words,voc_size) for words in corpus]
onehot_repr
```

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

#**Building LSTM model**

```python
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

#**Independent and dependent Features**

```python
X=np.array(embedded_docs)
y=np.array(message['lable'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=2)
```

#**Model training**

```python
### Finally Training
history=model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=20,batch_size=64)
```

```python
y_pred=(model.predict(X_test) > 0.5).astype("int32")
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
